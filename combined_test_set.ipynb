```python

# def extract_questions_and_answers(doc_path):
#     doc = docx.Document(doc_path)
#     questions = []
#     answers = []
#     chunk_numbers = []
   
#     # Variables to track the current question and answer
#     current_question = None
#     current_answer = []
#     current_chunk = None

#     # Iterate through paragraphs to extract questions, answers, and chunk numbers
#     for para in doc.paragraphs:
#         text = para.text.strip()
#         print(f"Paragraph: {para.text.strip()}")  # Debugging: Print every paragraph
  
#         # Check if the paragraph starts with a question (e.g., "1)" or "2)")
#         if text and text[0].isdigit() and text[1] == ')':
#             if current_question:  # Save previous question-answer pair
#                 answers.append("\n".join(current_answer).strip())
#                 chunk_numbers.append(current_chunk)
#                 current_answer = []
#             current_question = text
#         # If the paragraph contains an answer
#         elif "chunk_" in text.lower():
#             current_answer.append(text)  # The answer is part of the following paragraphs
#             chunk_numbers.append(text.split()[-1])  # Extract the chunk number from the last word
#         else:
#             current_answer.append(text)  # The answer might continue

 
#     # Append the last question-answer pair
#     if current_question:
#         answers.append("\n".join(current_answer).strip())
#         chunk_numbers.append(current_chunk)
    
#     # Print to debug
#     print(f"Extracted Questions: {questions}")
#     print(f"Extracted Answers: {answers}")
#     print(f"Extracted Chunk Numbers: {chunk_numbers}")

#     return questions, answers, chunk_numbers

# Function to extract questions, answers, and chunk numbers
def extract_questions_and_answers(doc_path):
    doc = docx.Document(doc_path)
    questions = []
    answers = []
    chunk_numbers = []

    current_question = None
    current_answer = []
    current_chunk = None

    for para in doc.paragraphs:
        text = para.text.strip()

        # Check if the paragraph contains "chunk_" (indicating an answer and chunk number)
        if "chunk_" in text.lower():
            if current_question:  # Save previous question and answer before moving to the next one
                answers.append("\n".join(current_answer).strip())
                chunk_numbers.append(current_chunk)
                current_answer = []
            current_answer.append(text)  # Add chunk information as the answer
            current_chunk = text.split()[-1]  # Extract chunk number (e.g., chunk_8)
        elif text.endswith('?'):  # Heuristic: Questions typically end with a "?"
            if current_question:  # Save previous question and answer before moving to the next one
                answers.append("\n".join(current_answer).strip())
                chunk_numbers.append(current_chunk)
                current_answer = []
            current_question = text
            questions.append(current_question)  # Store the question
        else:
            current_answer.append(text)  # Continue adding to the current answer

    # Add the last question-answer pair
    if current_question:
        answers.append("\n".join(current_answer).strip())
        chunk_numbers.append(current_chunk)

    # Debugging: Check the extracted lists
    print(f"Extracted Questions: {questions}")
    print(f"Extracted Answers: {answers}")
    print(f"Extracted Chunk Numbers: {chunk_numbers}")

    return questions, answers, chunk_numbers

```

```python
def load_paraphrased_questions(paraphrased_file):
    with open(paraphrased_file, "r", encoding="utf-8") as f:
        return [line.strip() for line in f.readlines() if line.strip()]
```

```python
def combine_into_test_set(original_questions, paraphrased_questions, answers, chunk_numbers, output_file="final_test_set.txt"):
    with open(output_file, "w", encoding="utf-8") as f:
        for i, (question, paraphrase, answer, chunk_num) in enumerate(zip(original_questions, paraphrased_questions, answers, chunk_numbers)):
            f.write(f"Q{i+1} (Chunk {chunk_num})\n")
            f.write(f"Original Question: {question}\n")
            f.write(f"Paraphrased Question: {paraphrase}\n")
            f.write(f"Answer: {answer}\n\n")

# === RUN ===
doc_path = "Manual QA.docx"
paraphrased_file = "paraphrased_questions.txt"

# Extract original questions, answers, and chunk numbers from the Word document
original_questions, answers, chunk_numbers = extract_questions_and_answers(doc_path)


# Load paraphrased questions from the .txt file
paraphrased_questions = load_paraphrased_questions(paraphrased_file)

# Combine everything into one final test set
combine_into_test_set(original_questions, paraphrased_questions, answers, chunk_numbers, output_file="final_test_set.txt")
print("Test set created successfully!")
```
